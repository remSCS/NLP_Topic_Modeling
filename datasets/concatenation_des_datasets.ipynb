## Import des librairies

```python
import pandas as pd
```

## Import Dataset

```python
file_list_train = []
file_list_test = []

for i in range(1,36):
    path = "C:/Users/julie/OneDrive/Documents/ESILV/Semestre_9/ML for NLP/Projet 2/Traduction avis clients/avis_"+str(i)+"_traduit.xlsx"
    df = pd.read_excel(path)
    if i != 25:
        if df["type"][1] != "test":
            file_list_train.append(path)
        if df["type"][1] == "test":
            file_list_test.append(path)
```

```python
# Créer une liste vide pour stocker les dataframes
df_list_train = []

# Utiliser une boucle pour lire chaque fichier Excel
for file in file_list_train:
  # Lire le fichier Excel en utilisant pandas
  df_train = pd.read_excel(file)
  # Ajouter le dataframe à la liste
  df_list_train.append(df_train)

# Concaténer tous les dataframes dans la liste en utilisant pandas
result_train = pd.concat(df_list_train)
```

```python
# Créer une liste vide pour stocker les dataframes
df_list_test = []

# Utiliser une boucle pour lire chaque fichier Excel
for file in file_list_test:
  # Lire le fichier Excel en utilisant pandas
  df_test = pd.read_excel(file)
  # Ajouter le dataframe à la liste
  df_list_test.append(df_test)

# Concaténer tous les dataframes dans la liste en utilisant pandas
result_test = pd.concat(df_list_test)
```
